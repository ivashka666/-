```python
from scipy import integrate
import numpy as np
import scipy.special as sc
from scipy.special import fresnel
import matplotlib.pyplot as plt
```

# Зависимость интенсивности от числа открытых зон френеля


```python
m = np.linspace(0, 10, 10000)
def I(m):         # x = 0
    S1, C1 = sc.fresnel(np.sqrt(2*m));
    S2, C2 = sc.fresnel(-np.sqrt(2*m));
    I = (C1-C2)**2+(S1-S2)**2;
    return I
I = [I(m) for m in m]

plt.plot(m, I)
plt.xlabel('m')
plt.ylabel('I(0, m)')
plt.title('Зависимость интенсивности от числа открытых зон френеля')
plt.grid()
plt.xlim(0, 10)
plt.xticks(np.arange(0, 10, 1))
```

# Распределение интенсивности для дифракционных картин с определенным числом открытых зон френеля

```python
x = np.linspace(-1, 1, 10000)
def Int(x, mfix):        
    S1full, C1full = sc.fresnel(np.sqrt(2*mfix)*(2*x/0.2 + 1));
    S2full, C2full = sc.fresnel(np.sqrt(2*mfix)*(2*x/0.2 - 1));
    Int = (C1full-C2full)**2+(S1full-S2full)**2;
    return Int
Int2 = [Int(x, 1.75) for x in x]
Int3 = [Int(x, 2.75) for x in x]
Int4 = [Int(x, 3.75) for x in x]
Int5 = [Int(x, 4.75) for x in x]
Int6 = [Int(x, 5.75) for x in x]
Int7 = [Int(x, 6.75) for x in x]

plt.plot(x, Int2)
plt.xlim(-0.2, 0.2)
plt.xlabel('x')
plt.ylabel('I(x)')
```

```python
FILENAME = "photo1.png" # имя файла
img0 = np.mean(imread(FILENAME), axis=2) # усредняем 3 канала чтобы получить grayscale изображение
plt.imshow(img0, cmap='gray');
```

```python
def calculate_intensity_array(array_name, array_shape, start_index, end_index):
    # Создание массива с заданными размерами
    intensity_array = np.zeros(array_shape[1])
    
    # Вычисление интенсивностей и заполнение массива
    for j in range(array_shape[1]):
        intensity = np.sum([array_name[i][j] for i in range(start_index, end_index + 1)]) / (end_index - start_index + 1)
        intensity_array[j] = intensity
    
    # Построение графика
    plt.plot(np.arange(array_shape[1]), intensity_array, color='red')
    plt.show()
    plt.savefig("figure1.png", dpi=400, transparent=True)
    return intensity_array
```
